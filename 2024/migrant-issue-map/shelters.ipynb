```python
#data comes from https://www.google.com/maps/d/u/0/viewer?mid=1iM2h_tzxmOCHp5Im2Epswg-FqpzdrTw&femb=1&ll=0%2C0&z=10

import fiona
import geopandas as gpd
file_path = 'shelters.kml'
fiona.drvsupport.supported_drivers['KML'] = 'rw'
# empty GeoDataFrame
gdf = gpd.GeoDataFrame()

# iterate over layers
for layer in fiona.listlayers(file_path):
  s = gpd.read_file(
    file_path, driver='KML', layer=layer)
  gdf = gdf.append(s, ignore_index=True)

gdf = gpd.GeoDataFrame(gdf)
```

```python
# clean up and add descriptions
gdf.columns = ['name','description','geometry']
gdf['name'] = gdf['name'].str.strip()
gdf['icon'] = 'lodging'
gdf['description'] = '<strong>' + gdf['name'] + '</strong>'
```
